# Adjusting the plots

```python
plt.plot(x,np.sin(x),color='blue')
plt.plot(x,np.sin(x+1),color='g') #shortcut
plt.plot(x,np.sin(x+2),color='0.8') #grey scale 
plt.plot(x,np.sin(x+3),color='#FFDD44') #hex value
plt.show()
```

```python
plt.plot(x,np.sin(x),color='blue',linestyle='solid')
plt.plot(x,np.sin(x+1),color='g',linestyle='dashed') 
plt.plot(x,np.sin(x+2),color='0.8',linestyle='dashdot')
plt.plot(x,np.sin(x+3),color='#FFDD44',linestyle='dotted')
plt.show()
```

```python
plt.plot(x,np.sin(x),color='blue',linestyle='-')
plt.plot(x,np.sin(x+1),color='g',linestyle='--') 
plt.plot(x,np.sin(x+2),color='0.8',linestyle='-.')
plt.plot(x,np.sin(x+3),color='#FFDD44',linestyle=':')
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g")
plt.plot(x,np.sin(x+1),'--c') 
plt.plot(x,np.sin(x+2),'-.k')
plt.plot(x,np.sin(x+3),':r')
plt.show()
```

```python
#limit for the x axis and y axis
plt.plot(x,np.sin(x),"-g")
plt.xlim(-1,12)
plt.ylim(-5,5)
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g")
plt.xlim(-1,12)
plt.ylim(-5,5)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("This is an Example for Title")
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g",label="Sin x")
plt.plot(x,np.sin(x+1),'--c', label='Sin(x+1)') 
plt.show()
```

```python
plt.plot(x,np.sin(x),"-g",label="Sin x")
plt.plot(x,np.sin(x+1),'--c', label='Sin(x+1)') 
plt.legend()
plt.show()
```
