# Pandas Series 

The pandas series is a one-dimensional labeled array capable of holding data of any type (integer, string, float, python objects, etc.). The axis labels are collectively called index.

```python
data = pd.Series([1,2,3,4,5,6])
data
```

```python
type(data) #one dimension array
```

```python
data.values #all values of data 
```

```python
data.index #index values of the data
```

```python
data[1]
```
